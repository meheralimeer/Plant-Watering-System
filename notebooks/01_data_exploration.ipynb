```python
import pandas as pd
import numpy as np

# Path to your CSV
file_path = "data/raw/original_dataset.csv"

# Load CSV
df = pd.read_csv(file_path)

print("Dataset loaded successfully!\n")
```

```python
# First 5 rows
print("\nFirst 5 rows:")
print(df.head())

# Shape
print("\nShape of dataset (rows, columns):")
print(df.shape)

# Column names
print("\nColumn names:")
print(df.columns)

# Data types
print("\nData types:")
print(df.dtypes)

# General info
print("\nDataset Info:")
print(df.info())
```

```python
print("\nStatistical Summary (Numerical Features):")
print(df.describe())

print("\nStatistical Summary (All Columns):")
print(df.describe(include='all'))
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())

print("\nTotal duplicate rows:")
print(df.duplicated().sum())
```

```python
target_col = "Plant_Health_Status"

print("\nClass Distribution:")
print(df[target_col].value_counts())

print("\nClass Distribution (Percentage):")
print(df[target_col].value_counts(normalize=True) * 100)
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
# covert timestamp into datetime format
```

```python
# days_since_last_watering feature
today = pd.to_datetime("today")

df["days_since_last_watering"] = (today - df["Timestamp"]).dt.days
```

```python
# Simple Moving Average (time-series logic)
df["watering_sma_3"] = df["days_since_last_watering"].rolling(window=3).mean()
```

```python
df.to_csv("data/processed/cleaned_data.csv", index=False)
print("Cleaned dataset saved to 'data/processed/cleaned_data.csv'")
```
